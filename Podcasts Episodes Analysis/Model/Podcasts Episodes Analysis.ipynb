# Visualisation of Data

```python
plt.figure(figsize=[20,30])
order= df['language'].value_counts().index
sns.countplot(data=df,y='language',order=order)
plt.title('Episodes by language',size=20);
```

```python
plt.figure(figsize=(20,12))
df['category'].value_counts().nlargest(20).plot.bar(color='red')
plt.gcf().autofmt_xdate()
plt.xlabel('category',size = 20)
```

```python
plt.figure(figsize=(20,12))
df['subcategory'].value_counts().nlargest(20).plot.bar(color='blue')
plt.gcf().autofmt_xdate()
plt.xlabel('subcategory',size = 20)
```

```python
plt.figure(figsize=(20,12))
df['author'].value_counts().nlargest(20).plot.bar(color='yellow')
plt.gcf().autofmt_xdate()
plt.xlabel('author',size = 20)
```

```python
plt.pie(df['explicit'].value_counts(), labels=df['explicit'].value_counts().index,counterclock=False,autopct='%.2f')
plt.title('Episodes BY explicit');
```
