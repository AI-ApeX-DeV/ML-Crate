# Cleaning Data

```python
data = dataset.drop_duplicates(subset='Name')
print('Total number of apps in the dataset = ', len(data))

dataset['Minutes'] = dataset['Minutes'].str.split(' ').str[0]
dataset.Minutes = dataset.Minutes.astype(int)

dataset['Votes'] = dataset['Votes'].str.replace(',', '')
dataset.Votes = dataset.Votes.astype(int)

print(data.head())
```

# EDA

## Bar Graph of Genres

```python
plt.figure(figsize=(20,10))
sns.countplot(data=data,x="genre",palette="icefire")
plt.xticks(fontsize=13,rotation=50)
plt.xlabel("Names of genres",fontsize=14)
plt.title("GENRE",fontsize=20,color="RED")
plt.show()
```

## Bar graph of ratings

```python
plt.figure(figsize=(20,9))
sns.countplot(data=data,x="Rating",palette="icefire")
plt.xticks(fontsize=13,rotation=50)
plt.xlabel("Ratings", fontsize = 15)
plt.ylabel("Count" , fontsize = 15)
plt.title("Ratings" , fontsize = 20, color = "BLUE")
plt.show()
```

## Pie chart of genres

```python
labels = dataset["genre"].value_counts().index
sizes = dataset["genre"].value_counts()
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99',"pink","yellow"]
plt.figure(figsize = (16,16))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',colors=colors,shadow=True, startangle=45)
plt.title('Genre',color = 'red',fontsize = 15)
plt.show()                                              
```

## Bar graph of runtime distribution

```python
plt.figure(figsize=(20, 12))
sns.histplot(dataset['Minutes'])
plt.title('Distribution of runtime per minutes')
plt.show()
plt.clf()
```

## Graph showing number of ratings of each movie

```python
order = dataset.groupby('Name').mean().sort_values('Votes', ascending = False).index

plt.figure(figsize=(20, 12))
ax = sns.barplot(y='Name', x='Votes', data = dataset, order=order)
ax.bar_label(ax.containers[0])
plt.title('Movie/TV Show with highest votes', fontsize = 20)
plt.show()
plt.clf()
```

## Details on number of votes

```python
print('The number of votes for all the movies combined is ' +  str(dataset.Votes.sum()) + ' votes')
print('The average vote in this dataset is ' + str(round(dataset.Votes.mean())) + ' vote per movie')
print('The minimum number of votes is ' + str(dataset.Votes.min()) + ' votes')
print('The maximum number of votes is ' + str(dataset.Votes.max()) + ' votes')
```

## Bar graph showing highest votes

```python
plt.figure(figsize=(20, 12))
ax = sns.barplot(y='genre', x='Votes', data = dataset, ci= None)
ax.bar_label(ax.containers[0])
plt.title('Genre with highest votes')
plt.show()
plt.clf()
```
