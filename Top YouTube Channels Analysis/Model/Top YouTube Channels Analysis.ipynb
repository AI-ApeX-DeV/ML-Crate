# Changing of DataType for Better Visualization of Data.

```python
df['video views '] = df['video views '].str.replace(',', '').astype(np.int64)
df['video views '] = pd.to_numeric(df['video views '])
```

```python
df.info()
```

```python
df.describe()
```

# Visualisation of Data

```python
plt.figure(figsize=(20, 10))
df.groupby('youtuber')['subscribers '].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('youtuber')['video views '].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 10))
df.groupby('youtuber')['video count '].mean().sort_values(ascending=False).plot(kind='bar')
```

# Plotting the Above data in Pie Chart

```python
plt.figure(figsize=(10, 10))
top = df['category '].value_counts().sort_values(ascending=False).head(13)
top.plot(kind='pie')
```

```python
plt.figure(figsize=(10, 10))
top = df['started '].value_counts().sort_values(ascending=False).head(13)
top.plot(kind='pie')
```
