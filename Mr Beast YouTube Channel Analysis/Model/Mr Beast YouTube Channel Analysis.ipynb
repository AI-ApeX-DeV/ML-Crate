# Data Visualization

```python
df.index=df.publishTime
```

```python
plt.figure(figsize=(21,6))
df.viewCount.plot()
plt.title("MR. BEAST YOUTUBE CHANNEL VIEWS")
plt.ylabel("VIEWS")
```

```python
plt.figure(figsize=(21,6))
df.likeCount.plot()
plt.title("MR. BEAST YOUTUBE CHANNEL LIKES")
plt.ylabel("LIKES")
```

```python
plt.figure(figsize=(21,6))
df.commentCount.plot()
plt.title("MR. BEAST YOUTUBE CHANNEL COMMENTS")
plt.ylabel("COMMENTS")
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(df.corr(),annot=True,linewidths=0.1)
```

```python
plt.figure(figsize=(21,6))
df.duration_seconds.plot()
plt.title("MR. BEAST YOUTUBE CHANNEL VIDEO DURATIONS")
plt.ylabel("VIDEO DURATIONS")
```

```python
plt.figure(figsize=(10, 5))
top = df['topicDetails.topicCategories'].value_counts().sort_values(ascending=False).head(8)
top.plot(kind='pie')
```
