### Encoding the categorical variables

```python
categorical = [col for col in prob_death.columns if prob_death[col].dtype == 'O']
categorical
```

```python
countries = prob_death[categorical]
countries
```

```python
country_data = pd.get_dummies(countries)
```

```python
country_data.head()
```

```python
prob_death.drop("Country", axis = 1, inplace = True)
```

```python
data_list = [country_data, prob_death]
```

```python
new_data = pd.concat(data_list, axis = 1)
new_data.head()
```

```python
new_data.shape
```

## Dropping columns with a high correlation

```python
drop_list = ['Male infant(0-1)', 'Female infant(0-1)', 'Male infants(0-5)', 'Female infants(0-5)']
```

```python
new_data.drop(drop_list, axis = 1, inplace = True)
new_data.head()
```

```python
x = new_data.drop(['Both genders (infants (0-1))', 'Both genders (infants (0-5))', 'neonatal_death'], axis = 1)
x.head()
```

```python
y1 = new_data['Both genders (infants (0-1))']
y2 = new_data['Both genders (infants (0-5))']
y3 = new_data['neonatal_death']
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
import pickle
```

### Separating the train and test variables

```python
X_train, X_test, y_train, y_test = train_test_split(x, y1, test_size=0.2)
```

```python
def predict(ml_model, dump):
    model = ml_model.fit(X_train,y_train)
    print('Training score : {}'.format(model.score(X_train,y_train)))
    y_prediction = model.predict(X_test)
    
    print('predictions are: \n {}'.format(y_prediction))
    print('\n')
    
    r2_score = metrics.r2_score(y_test,y_prediction)
    
    print('r2 score: {}'.format(r2_score))
    print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_prediction))
    print('Mean Squared Error:',metrics.mean_squared_error(y_test,y_prediction))
    print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test,y_prediction)))
    
    sns.distplot(y_test-y_prediction)
    
    if dump==1:
        ##dump your model using pickle so that we will re-use
        file = open("model_1.pkl",'wb')
        pickle.dump(model,file)
    
```

### Implementing the ML models on y1, y2, y3 variables

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

## Both Genders infants from age 0 to 1 years mortality Predictive analysis

```python
predict(LinearRegression(), 0)
```

```python
predict(DecisionTreeRegressor(), 1)
```

```python
predict(RandomForestRegressor(), 0)
```
