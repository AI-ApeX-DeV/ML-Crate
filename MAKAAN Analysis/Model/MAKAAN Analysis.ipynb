# Changing of DataType for Better Visualization of Data.

```python
df['Price_per_unit_area'] = df['Price_per_unit_area'].str.replace(',', '').astype(np.int64)
df['Price_per_unit_area'] = pd.to_numeric(df['Price_per_unit_area'])
```

```python
df['Price'] = df['Price'].str.replace(',', '').astype(np.int64)
df['Price'] = pd.to_numeric(df['Price'])
```

```python
df['Size'] = df['Size'].str.replace(' sq ft', '')
df['Size'] = df['Size'].str.replace(',', '').astype(np.int64)
```

```python
df.info()
```

```python
df.head()
```

# Visualisation of Data
