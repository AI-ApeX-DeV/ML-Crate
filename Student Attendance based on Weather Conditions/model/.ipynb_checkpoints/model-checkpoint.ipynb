```python
#dist of temperature
figure, axes = plt.subplots(1,2, figsize=(15,5), gridspec_kw={'width_ratios':[2,1]})
sns.histplot(data=data, x=data['Temperature'], kde=True, color='palegreen', ax=axes[0])
axes[0].set_ylabel('Frequency')
axes[0].grid(alpha=0.4)
sns.boxplot(data=data, y=data['Temperature'], color='palegreen', ax=axes[1])
axes[1].grid(alpha=0.4)
figure.suptitle('Temperature Distribution', fontsize=15)
plt.tight_layout(pad=1)
plt.show()
```

```python
# import required libraries for chi-square test
from scipy.stats.contingency import chi2_contingency, association

# visualize the student attendance on different weather conditions
plt.figure(figsize=(12,5))
sns.countplot(data=data, x=data['Weather'], hue=data['Attendance'], palette=['palegreen','orangered'])
plt.grid(alpha=0.4)
plt.title('Student Attedance on Different Weather Conditions', fontsize=15)
plt.show()

# display the observed value
contingency_table = pd.crosstab(data['Weather'], data['Attendance'])
print('Observed value: ')
display(contingency_table)

# perform Chi-square test and compute the effect size
res = chi2_contingency(contingency_table)
test_stat = round(res[0], 4)
pvalue = round(res[1], 4)
effect_size = round(association(contingency_table, method='cramer'), 4)
expected_freq = res[3]

# display the expected frequency
print('Expected frequency: ')
display(pd.DataFrame(expected_freq, index=contingency_table.index, columns=contingency_table.columns))

# print the test result
print('Test statistic: ', test_stat)
print('Test p-value: ', pvalue)
print('Effect size: ', effect_size)
```

```python
#since p value is lower than .05 (alpha), also effect size close to .5 indicates
#a good strength bw weather and student attendance
```

```python
# import essential libraries fo statistical test
from scipy.stats import ttest_ind

# visualization
figure, axes = plt.subplots(1,2, figsize=(15,5), gridspec_kw={'width_ratios':[1.5,1]})
sns.histplot(data=data, x=data['Temperature'], hue=data['Attendance'], kde=True, palette=['palegreen','orangered'], ax=axes[0])
axes[0].set_ylabel('Frequency')
axes[0].grid(alpha=0.4)
sns.boxplot(data=data, x=data['Attendance'], y=data['Temperature'], palette=['palegreen','orangered'], ax=axes[1])
axes[1].grid(alpha=0.4)
figure.suptitle('Present vs Absent: Temperature Distribution', fontsize=15)
plt.tight_layout(pad=1)
plt.show()

# display the summary statistic
groups = data.groupby('Attendance')
present = groups.get_group('Present')
absent = groups.get_group('Absent')
display(groups.Temperature.describe())

# perform indenpendent T-test to compare the mean temperature of boh groups
res = ttest_ind(present['Temperature'], absent['Temperature'])
test_stat = round(res[0], 4)
pvalue = round(res[1], 4)

# print the test result
print('Test statistic: ', test_stat)
print('Test pvalue: ', pvalue)
```

```python
# small test statistic and higher p value indicates that it doesn't have much immpact on attendance
```

```python
from sklearn.model_selection import train_test_split
```

```python
x,y=data.drop(columns=['Attendance','Temperature'],axis=1),data['Attendance']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
# One hot encoding
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder(drop='first')
Xtrain_encoded = onehot.fit_transform(x_train)
Xtest_encoded = onehot.transform(x_test)
```

```python
from sklearn.preprocessing import LabelEncoder

# instantiate the LabelEncoder
encoder = LabelEncoder()

# encode the target vector for both train and test dataset
ytrain_encoded = encoder.fit_transform(y_train)
ytest_encoded = encoder.transform(y_test)
```

```python
#Logistic Regression Implementation
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

# create a logistic regression Classifier
logreg = LogisticRegression(random_state=122, max_iter=5000)

# define the hyperparameter grid
param_grid = {
    'penalty': ['l1', 'l2'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'saga']
}

# create the GridSearchCV object
grid_search_logreg = GridSearchCV(logreg, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# fit the grid search to the data
grid_search_logreg.fit(Xtrain_encoded, ytrain_encoded)

# print the best parameters and the corresponding accuracy
print('Best Parameters: ', grid_search_logreg.best_params_)
print('Best Accuracy: ', grid_search_logreg.best_score_)

# get the best model
best_logreg = grid_search_logreg.best_estimator_
```

```python
# evaluate the best model on the test set
test_accuracy = best_logreg.score(Xtest_encoded, ytest_encoded)
print('Test Accuracy: ', test_accuracy)
```

```python
from sklearn.metrics import RocCurveDisplay
 
# check the ROC AUC curve
RocCurveDisplay.from_estimator(best_logreg, Xtest_encoded, ytest_encoded)
plt.show()
```

```python
# visualize the confusion matrix using a heatmap
from sklearn.metrics import confusion_matrix

ypred = best_logreg.predict(Xtest_encoded)

labels = y_test.unique()

cm = confusion_matrix(ytest_encoded, ypred, normalize='true')
plt.figure()
sns.heatmap(cm, annot=True, cmap='BuPu', xticklabels=labels, yticklabels=labels)
plt.show()
```

```python
# classification report
from sklearn.metrics import classification_report

report = classification_report(ytest_encoded, ypred)
print(report)
```

```python
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings('ignore')
regressor = RandomForestRegressor(n_estimators=10, random_state=0, oob_score=True)
regressor.fit(Xtrain_encoded,ytrain_encoded)
# Evaluating the model
from sklearn.metrics import mean_squared_error, r2_score
 
# Access the OOB Score
oob_score = regressor.oob_score_
print(f'Out-of-Bag Score: {oob_score}')
 
# Making predictions on the same data or new data
predictions = regressor.predict(Xtest_encoded)
 
# Evaluating the model
mse = mean_squared_error(ytest_encoded, predictions)
print(f'Mean Squared Error: {mse}')
 
r2 = r2_score(ytest_encoded, predictions)
print(f'R-squared: {r2}')
```

```python
#Decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

# Training with gini index
clf_gini = DecisionTreeClassifier(criterion="gini",random_state=100, max_depth=3, min_samples_leaf=5)
# Performing training
clf_gini.fit(Xtrain_encoded, ytrain_encoded)
y_predgini = clf_gini.predict(Xtest_encoded)

# Training with entropy
clf_entropy = DecisionTreeClassifier(criterion="entropy", random_state=100,max_depth=3, min_samples_leaf=5)
# Performing training
clf_entropy.fit(Xtrain_encoded, ytrain_encoded)
y_predentropy = clf_entropy.predict(Xtest_encoded)
```

```python
print('Accuracy using entropy: ',accuracy_score(ytest_encoded, y_predentropy))
print('Accuracy using gini: ',accuracy_score(ytest_encoded, y_predentropy))
```

```python
#Lasso Regression
# from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso
model = Lasso(alpha=1.0)
model.fit(Xtrain_encoded,ytrain_encoded)
predictions = model.predict(Xtest_encoded)
mse = mean_squared_error(ytest_encoded, predictions)
print("MSE:",mse)
```

```python
#Using Ridge
from sklearn.linear_model import Ridge 
model = Ridge(alpha =.1) 
model.fit(Xtrain_encoded, ytrain_encoded)
y_pred = model.predict(Xtest_encoded)
score = model.score(Xtest_encoded, ytest_encoded) 
print('Score:',score)
```

```python
#Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
gbm = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
gbm.fit(Xtrain_encoded,ytrain_encoded)
pred_y=gbm.predict(Xtest_encoded)
# accuracy
acc = accuracy_score(ytest_encoded, pred_y)
print("Gradient Boosting Classifier accuracy is : {:.2f}".format(acc))
```

```python
#XG Boost
import xgboost as xgb
xg_reg = xgb.XGBRegressor(objective ='reg:squarederror', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 100, random_state=42)
# Fit the model on the training data
xg_reg.fit(Xtrain_encoded, ytrain_encoded)

# Make predictions on the test set
predictions = xg_reg.predict(Xtest_encoded)

# Evaluate the model
mse = mean_squared_error(ytest_encoded, predictions)
print(f'Mean Squared Error: {mse}')
```

```python
#MLP
import tensorflow as tf
# Build an MLP model using Keras
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(Xtrain_encoded.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')  # 3 output neurons for the 3 classes in Iris dataset
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(Xtrain_encoded,ytrain_encoded, epochs=50, batch_size=32, validation_data=(Xtest_encoded,ytest_encoded), verbose=2)
```

```python
# Evaluate the model on the test set
y_pred = model.predict(Xtest_encoded)
# Evaluate the model on the test set
loss, accuracy = model.evaluate(Xtest_encoded, ytest_encoded, verbose=0)
print(f'Test Loss: {loss:.4f}')
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```
